```python
#1.準備
import pandas as pd
from maps import Maps

dataset1 = pd.read_csv("antenna.csv",index_col=[0])
dataset1
```

```python
#2-1.世界地図上のKaVAアンテナ分布

import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))

m = Maps()

## worldmapの引数
# DataFrame                 1.dataset, 
# アンテナの位置をプロット    2.color_mark="bo", 3.plot_size=15, 4.Antena_textsize=20,
# 図の大きさ、陸の色、海の色　5.conti_color="coral", 6.sea_color="aqua", 
# 緯、経線の間隔             7.longspan=5, 8.longfontsize=15, 9.latispan=5, 10.latifontsize=15,
# アンテナ文字表示,緯度,経度,地図の色　11.Antena_text_use=False, 12.long_use=True, 13.lati_use=True, 14.map_map_color=True 
# Basemapメソッドの引数      +Basemap(**keys)):

m.worldmap(dataset1,
          projection='cyl',llcrnrlat=20,urcrnrlat=50,llcrnrlon=120,urcrnrlon=150,resolution="l")
```

```python
#2-2.遠方の恒星から見た地球

import matplotlib.pyplot as plt
from datetime import datetime

plt.figure(figsize=(15,15))

m = Maps()

## from_starの引数
# DataFrame                 1.dataset, 
# 測定する対象時刻           3.observation_time(dataime型)
# 対象の恒星の赤経,赤緯　　　 4.Ra="00h00m00s", 5.Dec="00d00m00s"
# アンテナの位置をプロット    6.color_mark="bo", 7.plot_size=15, 8.Antena_textsize=8,
# 図の大きさ、陸の色、海の色　9.conti_color="coral", 10.sea_color="aqua", 
# 緯、経線の間隔             11.longspan=30, 12.latispan=30,
# アンテナ文字表示,緯度,経度,地図の色  13.Antena_text_use=False, 14.long_use=True, 15.lati_use=True, 16.map_map_color=True 

time = datetime(2017,11,17,20,33,0)

m.from_star(dataset1,time,"00h08m23.25988s","29d05m25.5520s")
```
